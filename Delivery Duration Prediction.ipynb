```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df=pd.read_csv(r'/Users/arijitguchhait/Desktop/Python/historical_data.csv')
df.head()
```

```python
from datetime import datetime
df['created_at']=pd.to_datetime(df['created_at'])
df['actual_delivery_time']=pd.to_datetime(df['actual_delivery_time'])
df['actual_total_delivery_durtion']=(df['actual_delivery_time']-df['created_at']).dt.seconds
df['estimated_non_prep_duration']=df['estimated_store_to_consumer_driving_duration']+df['estimated_order_place_duration']
df["busy_dashers_ratio"] = df["total_busy_dashers"] / df["total_onshift_dashers"]
df.info()
```

```python
df.isna().sum()
```

```python
print("# Percentange of missing value % :",df.isna().mean().mean()*100)
print("# Average order delivery time in minutes:",df['actual_total_delivery_durtion'].mean()/60)
```

```python
df.nunique()
```

```python
df['store_primary_category'].fillna(df['store_primary_category'].mode()[0],inplace=True)
```

```python
columns =['total_onshift_dashers','total_busy_dashers','total_outstanding_orders','estimated_store_to_consumer_driving_duration','estimated_non_prep_duration']
for i in columns:
    df[i]=df[i].fillna(df[i].mean())
```
